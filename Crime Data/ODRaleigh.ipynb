```python
#Dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt

#Raleigh Crime Data
url = "https://services.arcgis.com/v400IkDOw1ad7Yad/arcgis/rest/services/Police_Incidents/FeatureServer/0/query?where=1%3D1&outFields=crime_description,city,district,reported_year,reported_month,crime_category&outSR=4326&f=json"

response = requests.get(url).json()
pprint(response)
```

```python
#Creating data frame from pulled data
crimedata = response["features"]
formatted_data = [crime['attributes'] for crime in crimedata]
# pprint(formatted_data)

crimedata_pd = pd.DataFrame(formatted_data)
crimedata_pd.head(100)
```

```python
#Organizing crime descriptions into different columns
crimedata_pd["crime_description"]=crimedata_pd["crime_description"].str.split("/",1,False)
crimedata_pd.head()
```

```python
cat = []
descrip =[]
for index, row in crimedata_pd.iterrows():
    descrip.append(row["crime_description"][1])
    cat.append(row["crime_description"][0])
crimedata_pd["description"]=descrip
crimedata_pd["crime_cat"]=cat
crimedata_pd.drop(columns="crime_description",inplace=True)
crimedata_pd.head()
```

```python
#General categories
crimedata_pd["crime_cat"].value_counts()
```

```python
#Specific categories
crimedata_pd['description'].value_counts()
```

```python
#Cleaning up and merging similar categories
crimedata_pd['description'] = crimedata_pd['description'].replace({'Overdose':'Deceased Person','Suicide':'Deceased Person',
                                                                  'Overdose Death':'Deceased Person'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Missing Person (18 & over)':'Missing Person',
                                                                  'Runaway':'Missing Person', 'Missing Person (17 & under)': 'Missing Person'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Simple':'Assault','All Other':'Assault',
                                                                  'Aggravated':'Assault', 'Assault with an Object':'Assault'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Forcible Rape':'Sex crime', 'Forcible Fondling':'Sex crime',
                                                                  'Peeping Tom':'Sex crime','Forcible Sodomy':'Sex crime',
                                                                  'Statutory Rape':'Sex crime', 'Incest':'Sex crime'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Mental Commitment':'Other crimes','Found Property':'All Other Non-Offenses',
                                                                  'Lost Property':'Other crimes', 'Nonviolent (eg: neglect)':'Other crimes',
                                                                  'Injured Person':'Other crimes','Citizen Request':'All Other Non-Offenses',
                                                                  'Child Abuse or Neglect':'Other crimes','Truancy':'Other crimes'})
crimedata_pd['description'].value_counts()
```

```python
#Sorting dataframe by date
crimedata_pd = crimedata_pd.sort_values(['reported_year','reported_month'])
crimedata_pd = crimedata_pd.reset_index(drop=True)
crimedata_pd.head(50)
```

```python
#Removing non-crime data
crimes_pd = crimedata_pd.loc[crimedata_pd['description'] != 'All Other Non-Offenses']
crimes_pd = crimes_pd.reset_index(drop=True)
crimes_pd.head(50)
```

```python
#Grouping data by district
crimes_copy = crimes_pd.copy()
crimes_grouped_by_district = crimes_pd.groupby('district')
total_crimes_district = crimes_grouped_by_district['description'].count()
crime_by_district = crimes_grouped_by_district['description'].value_counts()
print(crime_by_district)
```

```python
#Crimes grouped by type
crimes_pd = crimes_pd.reset_index(drop=True)
crimes_grouped_by_type = crimes_copy.groupby('description')
total_crimes_type = crimes_grouped_by_type['district'].count()
crime_by_type = crimes_grouped_by_type['district'].value_counts()
print(crime_by_type)
```

```python
#Creating dataframes
district_pd = pd.DataFrame(crime_by_district)
district_pd = district_pd.rename(columns={'description':'count'})

type_pd = pd.DataFrame(crime_by_type)
type_pd = type_pd.rename(columns={'district':'count'})

print(district_pd)
print(type_pd)
```

```python
#Exporting data
district_pd.to_csv("district_crime.csv")
type_pd.to_csv("crime_type.csv")
```

```python
#Graphing by district
i = 0
districts = crimes_pd['district'].unique()
for district in districts:
    try:
        disdata = district_pd.loc[district]
        dis_graph = disdata.plot(kind = 'bar', legend = False, figsize = (6,6))
        plt.xlabel('Type of Crime')
        plt.ylabel('Total by District')
        plt.title(district)
        plt.grid(which='major',axis='y')
        plt.ylim(0,180)
        plt.tight_layout()
        plt.savefig(f"Crime_Graphs/districtgraph{i}.png")
        i += 1
    except:
        print("no graph")
```

```python
#Graphing by type
i = 0
types = crimes_pd['description'].unique()
for crimetype in types:
    try:
        typedata = type_pd.loc[crimetype]
        type_graph = typedata.plot(kind = 'bar', legend = False, figsize = (6,6))
        plt.xlabel('District')
        plt.ylabel('Total by Type')
        plt.title(crimetype)
        plt.grid(which='major',axis='y')
        plt.ylim(0,180)
        plt.tight_layout()
        plt.savefig(f"Crime_Graphs/typegraph{i}.png")
        i += 1
    except:
        print("no graph")
```
