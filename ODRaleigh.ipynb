```python
import requests
import json
from pprint import pprint
import pandas as pd

url = "https://services.arcgis.com/v400IkDOw1ad7Yad/arcgis/rest/services/Police_Incidents/FeatureServer/0/query?where=1%3D1&outFields=crime_description,city,district,reported_year,reported_month,crime_category&outSR=4326&f=json"

response = requests.get(url).json()
pprint(response)
```

```python
crimedata = response["features"]
formatted_data = [crime['attributes'] for crime in crimedata]
# pprint(formatted_data)

crimedata_pd = pd.DataFrame(formatted_data)
crimedata_pd.head()
```

```python
district_count = crimedata_pd['district'].value_counts()

cat_types = crimedata_pd['crime_category'].value_counts()
print(cat_types)
```
