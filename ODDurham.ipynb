```python
import requests
import json
from pprint import pprint
import pandas as pd

url = "https://webgis.durhamnc.gov/server/rest/services/PublicServices/Tables/MapServer/3/query?where=1%3D1&outFields=ADDRESS,CITY,ZIP,DOW,OFFENSE_DESCRIPTION&outSR=4326&f=json"

response = requests.get(url).json()
pprint(response)
```

```python
crimedata = response["features"]
formatted_data = [crime['attributes'] for crime in crimedata]
# pprint(formatted_data)

crimedata_pd = pd.DataFrame(formatted_data)
crimedata_pd.head(10)
```

```python
cat_types = crimedata_pd['OFFENSE_DESCRIPTION'].value_counts()
print(cat_types)
```
