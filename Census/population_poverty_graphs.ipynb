```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

filetoload = "Raleigh_zip_pop.csv"
```

```python
# loading file
data_df = pd.read_csv(filetoload)
data_df = data_df[['district', 'Population', 'Household Income', 'Poverty Rate']]
data_df
```

```python
# group by district and get mean values
data_grouped = data_df.groupby('district', as_index=False)
data_grouped.mean()
```

```python
#plot population in districts

plt.figure(figsize=(10,5))


plt.bar(data_df['district'], data_df['Population'], align="center")

plt.grid(axis='y', alpha=0.2)
plt.title("Total Population of each District")
plt.xlabel("District")
plt.ylabel("Population")

#Save graph
plt.savefig("Demographic Graphs/totalpop_district.png")

plt.show()
```

```python
#plot poverty rate in districts

plt.figure(figsize=(10,5))

plt.bar(data_df['district'], data_df['Poverty Rate'], align="center")

plt.grid(axis='y', alpha=0.2)
plt.title("Poverty Rate of each District")
plt.xlabel("District")
plt.ylabel("Poverty Rate (%)")

#Save graph
plt.savefig("Demographic Graphs/totalpov_district.png")

plt.show()
```
