```python
import requests
import json
from pprint import pprint
import pandas as pd

url = "https://www.chapelhillopendata.org/api/records/1.0/search/?dataset=police-incident-reports-written&rows=3000&sort=date_of_occurrence&facet=offense&facet=date_of_occurrence&facet=reported_as"

response = requests.get(url).json()
pprint(response)
```

```python
crimedata = response["records"]
formatted_data = [crime['fields'] for crime in crimedata]
# pprint(formatted_data)

crimedata_pd = pd.DataFrame(formatted_data)
crimedata_pd = crimedata_pd[{'city','street','date_of_occurrence','zipcode','reported_as','offense'}]
crimedata_pd.head(50)
```

```python
cat_types = crimedata_pd['reported_as'].value_counts()
print(cat_types)
```
